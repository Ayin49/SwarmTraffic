```python
import pandas as pd
import plotly.express as px
```

Analysing data from swarm control

```python
df = pd.read_xml(".\\stats\\tripinfo.xml")
```

keep track of waiting time, time loss, average speed

```python
fig = px.scatter(df, x="depart", y="waitingTime")
fig.show()
```

```python
df["waitingTime"].mean()
```

```python
fig = px.scatter(df, x="depart", y="timeLoss")
fig.show()
```

```python
df["timeLoss"].mean()
```

```python
# routeLength/duration
df["speed"] = df["routeLength"]/df["duration"]
fig = px.scatter(df, x="depart", y="speed")
fig.show()
```

```python
df["speed"].mean()
```
